```python
string="""this is just multi line string
and it will get printed"""
print(string)
print(string[0])
print(string[0:3])
print(string[2:])
print(string[:])
print(string[::-1])
print(len(string))
```

```python
string="u  AF"
print(string.lower())
print(len(string))
print(string.strip())
print(string.replace("U","PA"))
print(string.split())
print(string.find("F"))
print(string.count("F"))
print(string.startswith("F"))
print(string.endswith("F"))
print(string.title())
print(string.capitalize())
print(string.isdigit())
print(string.isalpha())
print(string.isupper())
print(string.islower())
```
