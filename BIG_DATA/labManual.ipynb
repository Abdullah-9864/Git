part 2: LOCAL SETUP

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master("local[2]") \
    .appName("bigdata") \
    .config("spark.driver.host", "127.0.0.1") \
    .config("spark.driver.bindAddress", "127.0.0.1") \
    .getOrCreate()

print("Ready:", spark.version, "Ready")
```

part 5: BASIC SUMMARY STAT

```python
fn.describe().show()
fn.select("gender").describe().show()
from pyspark.sql.functions import avg
fn.select(avg("assignments_completed")).show()
from pyspark.sql.functions import max
fn.select(max("assignments_completed")).show()
from pyspark.sql.functions import col, sum, when
fn.select([sum(when(col(c).isNull(), 1).otherwise(0)).alias(c)for c in fn.columns]).show()

```
