libraries

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as banana
```

Time vector

```python
t=np.linspace(0,1,1000)
```

Sine wave

```python
f=5
sin_signal=np.sin(2*np.pi*f*t)
```

Plot sine wave

```python
banana.plot(t,sin_signal)
banana.title("sine_wave-5Hz")
banana.xlabel("Time(s)")
banana.ylabel("Amplitude")
banana.grid(True)
banana.show()
```

Cosine wave

```python
cos_signal=np.cos(2*np.pi*f*t)
```

plot cosine wave

```python
banana.plot(t,cos_signal)
banana.title("cosine_wave-5Hz")
banana.xlabel("Time(s)")
banana.ylabel("Amplitude")
banana.grid(True)
banana.show()
```

Compare two frequencies

```python
signal_5 = np.cos(2 * np.pi * 5 * t)
signal_10 = np.sin(2 * np.pi * 10 * t)
banana.plot(t, signal_5, label="5 Hz")
banana.plot(t, signal_10, label="10 Hz")
banana.legend()
banana.grid(True)
banana.show()
```

Add noise

```python
noise = 1 * np.random.randn(len(t))
noisy_signal = cos_signal + noise
banana.plot(t, noisy_signal)
banana.title("Noisy Cosine Wave")
banana.xlabel("Time (s)")
banana.ylabel("Amplitude")
banana.grid(True)
banana.show()
```

Change Amplitude

```python
A = .5
amp_signal = A * np.cos(2 * np.pi * f * t)
banana.plot(t, amp_signal)
banana.title("Cosine Wave with Amplitude = 2")
banana.xlabel("Time (s)")
banana.ylabel("Amplitude")
banana.grid(True)
banana.show()
```

Multiple waves and superposition

```python
signal= np.sin(2* np.pi *5*t) + 0.5*np.sin(2* np.pi *10*t)
banana.plot(t, signal)
banana.show()
```

Fourier Transform

```python
freq = np.fft.fftfreq(len(signal), d=(t[1]-t[0]))
fft_signal = np.fft.fft(signal)
banana.plot(freq, np.abs(fft_signal))
banana.show()
```

3D wave

```python
X, Y = np.meshgrid(t, t)
Z = np.sin(2*np.pi*(X + Y))
```

plot 3D surface

```python
fig = banana.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
```

Labels

```python
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Amplitude')
ax.set_title('3D Sine Wave')

banana.show()
```
