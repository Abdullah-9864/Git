**Social Network Analysis (SNA)** is the study of how people, organizations, or objects interact. Instead of looking at individuals in isolation, SNA looks at the **structure** of their relationships.

### **1. Core Components**

**Nodes (Vertices):** The individual entities in the network (e.g., a person, a city, or a computer).
**Edges (Links):** The connections between nodes (e.g., a friendship, a road, or a data cable).

### **2. Types of Networks**

**Directed Network:** Relationships have a specific direction (e.g., Twitter—you follow someone, but they might not follow you back).

**Undirected Network:** Relationships are mutual (e.g., Facebook—if you are "Friends," the connection goes both ways).

**Weighted Network:** Edges have values representing "strength" or "distance" (e.g., a flight path where the weight is the number of passengers or the distance in miles).


### **3. Centrality (The "Importance" of a Node)**

Centrality measures identify the most important or influential nodes in a network based on their position.

**Degree Centrality:** Based on the number of direct connections. A person with many friends has high degree centrality.


**Closeness Centrality:** Based on how "near" a node is to everyone else. It measures how fast information can spread from that node.

**Betweenness Centrality:** Identifies "bridges." This node sits on the shortest path between other groups. If it is removed, communication between those groups breaks.

**Eigenvector Centrality:** Being connected to important people makes *you* important. It’s not just about how many people you know, but **who** you know.

**PageRank:** A specialized version of Eigenvector Centrality (used by Google). It ranks nodes based on the quantity and quality of incoming links.
