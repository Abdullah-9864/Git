***QUESTION 2: What would happen if you attempted to load a 2GB CSV file using Pandas on a system with 4GB RAM? Explain technically.***

Pandas loads the entire file into RAM at once, and a 2GB CSV expands to 4-6GB in memory due to Python object overhead. Since the system only has 4GB RAM, it will either throw a MemoryError or freeze completely. The operating system has no memory left to function.

***QUESTION 3: Compare inferSchema=True versus manually defining schema in Spark. Which is more efficient for large datasets and why?***

When using `inferSchema=True`, Spark scans the entire dataset twice — once to detect data types and once to load the data, which is very slow and memory-intensive on large datasets. Manual schema definition skips this scanning step entirely, allowing Spark to load data in a single pass, making it significantly faster and more efficient. Therefore, for large datasets, manual schema is always the recommended approach.

***QUESTION 4: Explain the concept of lazy evaluation in Spark. Provide an example from today’s lab.***


Lazy evaluation means Spark does **not execute any code immediately** when you write transformations — it only builds a plan and waits. Spark actually runs the code only when you call an **action** like `.show()`, `.count()`, or `.collect()`. For example, in today's lab when we wrote `df = spark.read.csv(...)` and `df.select(avg("salary"))`, nothing actually executed until we called `.show()` at the end — that single `.show()` triggered all the previous steps at once, making it more efficient by optimizing the entire plan before running.
